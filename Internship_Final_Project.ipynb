Calculate Hospital Stay Days

```python
df['Admission_Date']=pd.to_datetime(df['Admission_Date'])
df['Discharge_Date']=pd.to_datetime(df['Discharge_Date'])
df['Stay_Days']=(df['Discharge_Date']-df['Admission_Date']).dt.days

df[['Patient_ID','Stay_Days']].head()
```

 Average Hospital Stay


```python
df['Stay_Days'].mean()
```

Age Group Analysis

```python
df['Age_Group']=pd.cut(df['Age'],
                       bins=[0,18,40,60,100],
                       labels=['Child','Young','Adult','Senior'])

df['Age_Group'].value_counts()
```

Long Stay Classification

```python
def stay_type(days):
    if days<=3:
        return "Short"
    elif days<=7:
        return "Medium"
    else:
        return "Long"

df['Stay_Type']=df['Stay_Days'].apply(stay_type)
df[['Patient_ID','Stay_Days','Stay_Type']].head()
```

Monthly Admissions

```python
df['Admission_Month']=df['Admission_Date'].dt.month
df['Admission_Month'].value_counts().sort_index()
```
